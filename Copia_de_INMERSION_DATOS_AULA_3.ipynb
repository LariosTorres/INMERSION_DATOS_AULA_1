```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
tipos_inmuebles = inmuebles.query("Tipo in ['Apartamento', 'Casa', 'Local']")
grafica = sns.histplot(data=tipos_inmuebles, x = 'Precio_Millon', kde= True, hue = "Tipo", bins = 2000, palette = "YlOrBr")
grafica.set_title("Comparacion de precios de inmuebles en Bogota", fontsize = 15)
grafica.set_xlabel("Numero de inmuebles en venta", fontsize = 12)
grafica.set_ylabel("Precio de inmuebles en Bogota", fontsize = 12)
plt.ylim((0,300))
plt.xlim((70,500))
plt.show()
```

```python
precio_m2_todos = inmuebles.Precio.astype(float)/(inmuebles.Área.astype(float)*100)
inmuebles['precio_m2'] = precio_m2_todos

barrios_m2 = inmuebles.groupby('Barrio')['precio_m2'].mean().reset_index().sort_values(by='precio_m2', ascending=False).head(10)

plt.figure(figsize = (15,10))
grafica = sns.barplot(data=barrios_m2, y='Barrio', x='precio_m2', color='Purple')
grafica.set_title("Precios por metro cuadrado segun el barrio", fontsize = 15)

plt.xlabel("Precio por metro cuarado", fontsize = 15)
plt.ylabel("Barrio", fontsize = 15)

plt.savefig("/content/drive/MyDrive/precio_m2_barrio.png", format="png")
plt.show()
```

```python
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Área'] 
inmuebles.head(3)
```

```python
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Área']
datos_barrio
```

```python
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```
