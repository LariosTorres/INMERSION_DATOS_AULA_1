```python
#Importamos la librería pandas para poder analizar los datos, a su vez, le asignamos el nombre de pd 
#con el cual le daremos un nombre corto y sencillo de recordar, aunque usar el "as pd" es opcional
#es recomendable su uso para una mejor compresión de lo que se está haciendo
import pandas as pd

#Creamos una variable llamada "inmuebles y le asignamos un valor, en este caso la variables llama a la 
#libreria pandas, la cual le dimos el nombre de "pd" y le iremos que lea el archivo csv que queremos analizar
inmuebles = pd.read_csv("/content/drive/MyDrive/inmuebles_bogota.csv")

#Mostramos la "cabeza" del archivo csv, el cual fue guardado en la variable "inmuebles" 
#lo cual consiste en mostrar los 5 primero elementos del archivo
inmuebles.head()
```

```python
#Creacion de un diccionario llamado "columnas" en el cual se le asignan un valor a un nombre
columnas = {'Baños':'Banos','Área':'Area'}
#Usando la variable inmueble se le renombrara ciertos nombres de los atributos del archivo
#Los atributos que serán afectados solo serán los que coincidan con el nombre dado en el diccionario de columnas
#Dando como resultado que el atributo "Baños" cambie a "banos" y que el atributo "Área" cambie a "Area"
inmuebles = inmuebles.rename(columns=columnas)
#Una vez echo lo anterior se puede usar sample() para mostrar como quedo la "modificación" se debe recordar que
#las "modificaciones" solo serán hechas al mostrar la información, además que uno puede escoger el número de
#registros que se mostraran de forma aleatoria
inmuebles.sample(10)
```
